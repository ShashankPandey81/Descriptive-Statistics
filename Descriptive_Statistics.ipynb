Q1: Understanding Central Tendency (Easy)
A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?

- Given data of sales of muffins 10, 11, 12, 12, 13, 14, 15

Mean = 10 + 11 + 12 + 12 + 13 + 14 + 15  = 87

Mean = 87 / 7 = 12.42

Median = 12

Mode = 12

The most representative value is 12 dozens because it occurs most frequently and lies at the center of the data, making it a good indicator of typical daily sales
